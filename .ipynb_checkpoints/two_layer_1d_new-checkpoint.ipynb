```python
import numpy as np 
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as FF
#from flow import NormalizingFlow, fully_connected
from losses import Relative_Entropy_FC, Linear_loss,JKO_loss_a, JKO_loss_l, Porous_medium_loss,Interaction_loss
#from two_d_test_dist import double_banana, sine, star_gaussian, Mixture_Gaussian
import warnings
warnings.filterwarnings("ignore")
from OU_process import OU_process, OU_parameters, OU_unif, OU_exp
from two_layer import two_layer
from densities import Porous_medium
from SDE_sampler import sde_sampler
from compute_G import G_matrix,G_matrix_b,G_matrix_w
from FPE import FPE
import numpy.polynomial.hermite as hermit 


# set seed 
seed_num = 154
torch.manual_seed(seed_num) #154 #1099
np.random.seed(0)


# number of particles 
num_particles = int(1e6)
quadrature = False
if quadrature:
    num_particles = 100
num_particles_final = int(1e6)
Y_size = int(num_particles)

dimension = 1

############################################################
# OU process parameters
############################################################
# dX_t = -theta * (X_t - mu) dt + sigma dW_t
OU = OU_parameters()
theta_ou,sigma_ou,mu_ou  = OU.parameters()
D = sigma_ou**2/2
shift = 1

############################################################
# choice of potential
############################################################
choice = 1
def V(x,dimension = dimension):
    if choice == 0: return 0
    if choice == 1:
        # quadraticï¼Œthis is also an OU process 
        return theta_ou * (x-mu_ou)**2/2
    if choice == 2:
        if quadrature:
            return (x-shift/np.sqrt(2))**6/6
        else:
            return (x-shift)**6/6
    if choice == 3:
        try:
            return torch.abs(x)*4
        except:
            return np.abs(x)*4
    if choice == 4:
        return (x-shift)**4/4-(x-shift)**2/2+(x-shift)/8
    if choice == 5:
        return x**6/6 - x**5/3 + x**4/10-x/15
    if choice == 6:
        if quadrature:
            return (x-shift/np.sqrt(2))**4/4-(x-shift/np.sqrt(2))**2/2
        else:
            return (x-shift)**4/4 -(x-shift)**2/2
        
def grad_V(x,dimension = dimension):
    if choice == 0: return 0
    if choice == 1:
        return theta_ou*(x-mu_ou)
    if choice ==2:
        if quadrature:
            return (x-shift/np.sqrt(2))**5
        else:
            return (x-shift)**5
    if choice == 3:
        try:
            return (x)/torch.abs(x)*4
        except:
            return (x)/np.abs(x)*4
    if choice == 4:
        return (x-shift)**3-(x-shift)+1/8
    if choice == 5:
        return x**5-x**4*5/3+x**3*4/10-1/15
    if choice == 6:
        if quadrature:
            return (x-shift/np.sqrt(2))**3-(x-shift/np.sqrt(2)) 
        else:
            return (x-shift)**3 -(x-shift) 


############################################################
# Interaction potential
############################################################
chi = .5
def W(x):
    return torch.log(torch.abs(x))*2*chi

############################################################
# reference distribution: Gaussian, uniform, exponential, porous media
############################################################
ref_dist = 1
ref_mean, ref_var = 0.0, 1
unif_bound = 1
lam = 1
t0=1
C = (np.sqrt(3)/8)**(2/3)
x_upper = np.sqrt(12 * C *t0**(2/3))
x_lower = -x_upper
po = Porous_medium(t0=t0)
dimension = 1 # we are only considering 1D problems in this paper

def sample_X(ref_dist,num_particles=num_particles):
    weights = torch.zeros((num_particles,dimension))
    ## Gaussian
    if ref_dist == 1:
        if quadrature:
            pos,weights = hermit.hermgauss(num_particles)
            X = np.sqrt(2)*torch.tensor(pos).view(num_particles,dimension)
            weights = torch.tensor(weights).view(num_particles,dimension)
        else:
            X = torch.normal(ref_mean, std = np.sqrt(ref_var), size=(num_particles,dimension))
        X_final = torch.normal(ref_mean, std = np.sqrt(ref_var), size=(num_particles_final,dimension))
    ## uniform
    elif ref_dist == 2:
        X = torch.linspace(-unif_bound,unif_bound,num_particles)
        X = X.view((num_particles,dimension))
        X_final = torch.linspace(-unif_bound,unif_bound,num_particles_final)
        X_final = X_final.view((num_particles_final,dimension))
    ## exponential
    elif ref_dist == 3:
        X1 = torch.empty(int(num_particles/2),dimension)
        X1.exponential_(lam)
        X = torch.cat((-X1,X1),0)
        X1 = torch.empty(int(num_particles_final/2),dimension)
        X1.exponential_(lam)
        X_final = torch.cat((-X1,X1),0)
    ## porous media (Barenblatt profile)
    elif ref_dist == 4:
        X = torch.tensor(po.sampling(num_particles=num_particles))
        X_final = torch.tensor(po.sampling(num_particles=num_particles_final))
        X = X.view((num_particles,dimension))
        X_final = X_final.view((num_particles_final,dimension))
    return X,X_final,weights

X,X_final,weights = sample_X(ref_dist=ref_dist)
#X = X.to(dtype=torch.float64)
weights= weights.to(dtype=torch.float64)/np.sqrt(np.pi)
#weights = weights/np.sqrt(np.pi)
#XX,_ = sample_X(ref_dist=ref_dist)

# pdf of standard gaussian
def gauss(x,mean=ref_mean,var=ref_var,dim=1): 
    gauss = 1/(2*np.pi)**(dim/2)/np.sqrt(var)*np.exp(-(x-mean)**2/2/var) 
    return gauss

def laplace(x,lam_=lam):
    lap = 1/2 * lam_ * np.exp(-lam_*np.abs(x))
    return lap

def uniform(x,unif_bound=unif_bound):
    return 1/unif_bound/2 * (np.abs(x)<=unif_bound)

def porous(x):
    return po.density(x=x,t=0)

def initial_dist(x):
    if ref_dist == 1:
        return gauss(x)
    if ref_dist == 2:
        return uniform(x)
    if ref_dist == 3:
        return laplace(x)
    if ref_dist == 4:
        return porous(x)

############################################################
# solution of the linear transport equation
############################################################
# dp/dt = p * V'' + dp/dx * V' where V(x) = theta_ou * (x-mu_ou)**2/2
# and p(x,0) = gauss(x)
def linear_pde(x,t,mu=mu_ou,theta=theta_ou):
    arg = (x - mu)*np.exp(t*theta)+mu
    return gauss(arg)*np.exp(t*theta)

def safe_log(z):
    return torch.log(z + 1e-7)

relative_entropy_loss = Relative_Entropy_FC(potential = V, dimension = dimension,D=D,weights=weights)
just_entropy_loss = Relative_Entropy_FC(potential=0, dimension = dimension, D=D,weights=weights )
linear_loss = Linear_loss(potential=V,dimension=dimension,weights=weights)
porous_medium_loss = Porous_medium_loss(m=2)
interaction_loss = Interaction_loss(W=W,n=num_particles)
entropy = 1
implicit = False
interval = 1000
m_list = np.array([2,4,8,12,16])
m_list = np.array([4,9])
#m_list = np.array([100])
m_list = np.array([4,8,16,32])
m_list = np.array([4])

lr =2e-5
final_T = 1e-1
Max_Iter = int(np.ceil(final_T/lr)) + 1
output_history_list = torch.zeros([len(m_list),int((Max_Iter-1)/interval)+1,num_particles_final])
x_size = int(1e6)
x_axis = np.linspace(-8,8,x_size).reshape(x_size,1)
y_axis_final = np.zeros((len(m_list),np.shape(x_axis)[0]))
det_jac_final = np.zeros((len(m_list),np.shape(x_axis)[0]))
early_stopping = True
e_stopping_tol = 1e-7
update_inner_X = False
update_outer_X = False
segment = 10
w_length = 0
tol = 1e-4
LM = False
use_full_G = True 
eps = 2e-6
    
for index,m in enumerate(m_list):
    ############################################################
    # optimization parameters
    ############################################################
    #optimizer = torch.optim.Adam(flow.parameters(), lr=lr, betas=(0.9, 0.999), eps=1e-08)
    #lr = 0.00005*m
    factor = 0.00005
    inner_lr = factor #* m 
    h = lr
    Inner_max_iter = 10000
    Max_Iter = int(np.ceil(final_T/lr)) + 1
    output_history_list = torch.zeros([len(m_list),int((Max_Iter-1)/interval)+1,num_particles_final])
    y_axis = np.zeros((int((Max_Iter-1)/interval)+1,np.shape(x_axis)[0]))
    #y_axis = torch.zeros(x_axis.size(0))
    fully_con = two_layer(m = m,bias_init=ref_mean)
    aux_fc = two_layer(m = m,bias_init=ref_mean)
    #aux2_fc = two_layer(m = m,bias_init=ref_mean)
    init_fc = two_layer(m = m,bias_init=ref_mean)
    init_fc.load_state_dict(fully_con.state_dict())
    optimizer = torch.optim.SGD(fully_con.parameters(),lr = lr*(m)**2 )
    #aux_optimizer = torch.optim.SGD(aux2_fc.parameters(),lr=lr)
    inner_optimizer = torch.optim.Adam(fully_con.parameters(),lr = inner_lr)
    #scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, lrdecay)
    if entropy == 1:
        loss_jko = JKO_loss_a(potential=V, stepsize=h, dimension=dimension,D=D)
    elif entropy == 0:
        loss_jko = JKO_loss_l(potential=V, stepsize=h, dimension=dimension)
    
    
    error = np.zeros(Max_Iter)
    out_history_ = torch.zeros([int((Max_Iter-1)/interval)+1,num_particles_final])
    b_count = 0

    for k in range(int(np.ceil(final_T/lr))+1 ):
        optimizer.zero_grad()
        if k%interval ==0:
            out = fully_con(X_final)
            out = out.view(-1)
            out_history_[int(k/interval),:] = out
            y_axis[int(k/interval),:] = fully_con(torch.tensor(x_axis)).detach().numpy().reshape(-1)
            
        if update_outer_X:
            X,_ = sample_X(ref_dist=ref_dist)
        zk = fully_con(X) 
        
        if entropy == 1: 
            det_jac= fully_con.det_jac(X)
            my_loss = relative_entropy_loss(zk=zk,det_jacobian=det_jac)
        elif entropy == 0:
            my_loss = linear_loss(zk)
        elif entropy == 2:
            det_jac = fully_con.det_jac(X)
            pz = po.density(x=X,t=0).view(num_particles)
            my_loss = porous_medium_loss(det_jacobian=det_jac,pz=pz)
        elif entropy == 3:
            #zkk = fully_con(XX)
            det_jac= fully_con.det_jac(X)
            #my_loss = interaction_loss(zk.view(-1,1),zkk.view(-1,1),det_jacobian=det_jac)
            my_loss = interaction_loss(zk.view(-1,1),det_jacobian=det_jac)
        my_loss.backward(retain_graph=True) 

        aux_fc.zero_grad()
        aux_fc.load_state_dict(fully_con.state_dict())
        old_loss = torch.tensor(float('inf'))
        if update_inner_X:
            X,_ = sample_X(ref_dist=ref_dist)
        if implicit == True: 
            for j in range(Inner_max_iter):
                #scheduler.step()
                #X = torch.normal(ref_mean, std = np.sqrt(ref_var), size=(num_particles,dimension))
                s_0  = aux_fc(X)
                s_1 = fully_con(X)
                inner_optimizer.zero_grad()
                if entropy == 1:
                    det_jacobian = torch.abs(fully_con.det_jac(X))
                    log_jac = safe_log(det_jac)                    
                    inner_loss = loss_jko(s_0, s_1,log_jac)
                else:
                    inner_loss = loss_jko(s_0, s_1)
                inner_loss.backward(retain_graph=True)
                inner_optimizer.step()
                if j <=2 or Inner_max_iter-j <=2:
                    print(f'{j}: inner_loss = {inner_loss.data}, b_grad = {torch.max(torch.abs(fully_con.bias.grad.data))}, w_grad = {torch.max(torch.abs(fully_con.w.grad.data))}')
                if early_stopping:
                    if torch.max(torch.abs(fully_con.w.grad.data)) <= e_stopping_tol: #/m:
                        print(f'{j}: break. b_grad = {torch.max(torch.abs(fully_con.bias.grad.data))}, w_grad = {torch.max(torch.abs(fully_con.w.grad.data))}')
                        break          
        else:
            # elif update_inner_X:
            #     Y,_ = sample_X(ref_dist=ref_dist)
            # else:
            #     #Y = X
            #     Y = torch.linspace(-5,5,num_particles).view(num_particles,1)
            if use_full_G:
                # if quadrature:
                #     Y = X
                #     dfdb = aux_fc.df_db(Y)
                #     dfdw = aux_fc.df_dw(Y)
                #     for i in range(Y_size):
                #         gradient_b = dfdb[i,:].view(2*m,1)
                #         gradient_w = dfdw[i,:].view(2*m,1)
                #         gradient = torch.cat((gradient_b,gradient_w))
                #         if quadrature:
                #             if i == 0:
                #                 agg_gram = torch.matmul(gradient,torch.t(gradient))*weights[i]
                #             else:
                #                 agg_gram += torch.matmul(gradient,torch.t(gradient))*weights[i]
                #         else:
                #             if i == 0:
                #                 agg_gram = torch.matmul(gradient,torch.t(gradient))*gauss(Y[i])
                #             else:
                #                 agg_gram += torch.matmul(gradient,torch.t(gradient))*gauss(Y[i])
                # else: 
                #     agg_gram = G_matrix(w = fully_con.w.data[0:m],wtilde=fully_con.w.data[m:],b=fully_con.bias.data[0:m],btilde=fully_con.bias.data[m:],m=m)

                agg_gram = G_matrix(w = fully_con.w.data[0:m],wtilde=fully_con.w.data[m:],b=fully_con.bias.data[0:m],btilde=fully_con.bias.data[m:],m=m)

                # #agg_gram = agg_gram/num_particles
                # #agg_gram = agg_gram * (10/num_particles)
                #agg_gram = G_matrix(w = fully_con.w.data[0:m],wtilde=fully_con.w.data[m:],b=fully_con.bias.data[0:m],btilde=fully_con.bias.data[m:],m=m)
                if LM == True:
                        agg_gram = agg_gram + torch.eye(m*4)*tol
                # #print(torch.linalg.cond(agg_gram))
                old_grad_w = fully_con.w.grad.data
                
                if entropy == 0: 
                    old_grad_b = fully_con.bias.grad.data
                elif entropy == 1:
                    old_grad_b = torch.zeros(2*m)
                    #arg1 = fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1))/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1) - eps)
                    #arg1 = torch.abs(arg1)
                    #log1 = torch.log(arg1)
                    #nan_mask1 = torch.isnan(log1)
                    #log1[nan_mask1]=0
                    old_grad_b[0:m] = fully_con.bias.grad.data[0:m] + initial_dist(fully_con.bias.data[0:m])\
                        *torch.log(fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1)  )/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1) - eps))

                    #arg2 = fully_con.det_jac(fully_con.bias.data[m:].view(-1,1))/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1) + eps)
                    #arg2 = torch.abs(arg2)
                    #log2 = torch.log(arg2)
                    #nan_mask2 = torch.isnan(log2)
                    #log2[nan_mask2]=0   
                    old_grad_b[m:] = fully_con.bias.grad.data[m:] - initial_dist(fully_con.bias.data[m:])\
                        *torch.log(fully_con.det_jac(fully_con.bias.data[m:].view(-1,1)  )/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1) + eps ))  
                else:
                    old_grad_b = torch.zeros(2*m)
                    old_grad_b[0:m] =  initial_dist(fully_con.bias.data[0:m])**2\
                        *(1/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1)-eps )-1/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1)))
                    old_grad_b[m:] =  -initial_dist(fully_con.bias.data[m:])**2\
                        *(1/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1) +eps)-1/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1)))
                
                old_grad = torch.cat((old_grad_b,old_grad_w))
                #old_grad = old_grad.to(dtype=agg_gram.dtype)
                new_grad = torch.linalg.lstsq(agg_gram,old_grad).solution
                #new_grad = new_grad.to(dtype=torch.float64)
                #inv_ = torch.linalg.pinv(agg_gram)
                #new_grad = torch.matmul(inv_,old_grad)
                #new_grad = torch.linalg.solve(agg_gram,old_grad)
                fully_con.bias.grad = new_grad[:2*m]
                fully_con.w.grad = new_grad[m*2:]


            else:
                #only update w
                if k%segment<w_length or b_count>=100000:
                    if quadrature:
                        Y = X
                        dfdw = aux_fc.df_dw(Y)
                        for i in range(Y_size):
                            gradient_w = dfdw[i,:].view(2*m,1)
                            gradient = gradient_w
                            if quadrature:
                                if i == 0:
                                    agg_gram = torch.matmul(gradient,torch.t(gradient))*weights[i]
                                else:
                                    agg_gram += torch.matmul(gradient,torch.t(gradient))*weights[i]
                            else:
                                if i == 0:
                                    agg_gram = torch.matmul(gradient,torch.t(gradient))*gauss(Y[i])
                                else:
                                    agg_gram += torch.matmul(gradient,torch.t(gradient))*gauss(Y[i])
                    else:
                    # #agg_gram = agg_gram/num_particles
                    # agg_gram = agg_gram * (10/num_particles)
                        agg_gram = G_matrix_w(b=fully_con.bias.data[0:m],btilde=fully_con.bias.data[m:],m=m)
                    if LM == True:
                        agg_gram = agg_gram + torch.eye(m*2)*tol
                    #print(f'w condition: {torch.linalg.cond(agg_gram)}')
                    old_grad_w = fully_con.w.grad.data
                    old_grad = old_grad_w
                    #old_grad = old_grad.to(dtype=agg_gram.dtype)
                    new_grad = torch.linalg.lstsq(agg_gram,old_grad).solution
                    #new_grad = new_grad.to(dtype=torch.float32)
                    #new_grad = torch.linalg.solve(agg_gram,old_grad)
                    fully_con.w.grad = new_grad
                    fully_con.bias.grad = torch.zeros(2*m)


                ## only update b 
                else:
                    
                    if quadrature:
                        Y = X
                        dfdb = aux_fc.df_db(Y)
                        for i in range(Y_size):
                            gradient_b = dfdb[i,:].view(2*m,1)
                            gradient = gradient_b
                            if quadrature:
                                if i == 0:
                                    agg_gram = torch.matmul(gradient,torch.t(gradient))*weights[i]
                                else:
                                    agg_gram += torch.matmul(gradient,torch.t(gradient))*weights[i]
                            else:
                                if i == 0:
                                    agg_gram = torch.matmul(gradient,torch.t(gradient))*gauss(Y[i])
                                else:
                                    agg_gram += torch.matmul(gradient,torch.t(gradient))*gauss(Y[i])
                    else: 
                        
                    # #agg_gram = agg_gram/num_particles
                    # agg_gram = agg_gram * (10/num_particles)
                        agg_gram = G_matrix_b(w=fully_con.w.data[0:m],wtilde=fully_con.w.data[m:],b=fully_con.bias.data[0:m],btilde=fully_con.bias.data[m:],m=m)
                    if LM == True:
                        agg_gram = agg_gram + torch.eye(m*2)*tol
                    #
                    if entropy == 0:
                        old_grad_b = fully_con.bias.grad.data
                    elif entropy == 1:
                        old_grad_b = torch.zeros(2*m)
                        old_grad_b[0:m] = fully_con.bias.grad.data[0:m] + initial_dist(fully_con.bias.data[0:m])\
                            *torch.log(fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1)  )/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1) - eps))
                        old_grad_b[m:] = fully_con.bias.grad.data[m:] - initial_dist(fully_con.bias.data[m:])\
                            *torch.log(fully_con.det_jac(fully_con.bias.data[m:].view(-1,1)  )/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1) + eps ))
                    else:
                        old_grad_b = torch.zeros(2*m)
                        old_grad_b[0:m] =  -initial_dist(fully_con.bias.data[0:m])**2\
                            *(1/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1)+1e-6 )-1/fully_con.det_jac(fully_con.bias.data[0:m].view(-1,1)))
                        old_grad_b[m:] =  -initial_dist(fully_con.bias.data[m:])**2\
                            *(1/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1) - 1e-6)-1/fully_con.det_jac(fully_con.bias.data[m:].view(-1,1)))
                    old_grad = old_grad_b
                    old_grad = old_grad.to(dtype=agg_gram.dtype)
                    new_grad = torch.linalg.lstsq(agg_gram,old_grad).solution
                    #new_grad = torch.linalg.solve(agg_gram,old_grad)
                    fully_con.bias.grad.data = new_grad
                    fully_con.w.grad.data = torch.zeros(2*m)
                    b_count = b_count + 1


        if k%interval ==0:
            if implicit == 1:
                print(f'inner loss: {inner_loss}')
            print(f'm={m}, outer iteration: {k}/{int(np.ceil(final_T/lr))}, outer loss: {my_loss}' )

            
        
        #error[k] = my_loss
        optimizer.step()
        #fully_con.w.data[0:m] = torch.clamp(fully_con.w.data[0:m], min=-0.)
        #fully_con.w.data[m:] = torch.clamp(fully_con.w.data[m:], max=0.)
        # ind1 = torch.nonzero(fully_con.w.data[0:m]  < 0)
        # ind2 = torch.nonzero(fully_con.w.data[m:]  > 0)
        # fully_con.w.data[0:m][ind1] = aux_fc.w.data[0:m][ind1]
        # fully_con.w.data[m:][ind2] = aux_fc.w.data[m:][ind2]

    if out_history_[0,0] != 0:    
        out_history = out_history_
    output_history_list[index,:,:] = out_history_
    #y_axis_final[index,:] = y_axis[-1]
    y_axis_final[index,:] = fully_con(torch.tensor(x_axis)).detach().numpy().reshape(-1)
    det_jac_final[index,:] = fully_con.det_jac(torch.tensor(x_axis)).detach().numpy().reshape(-1)
    
```

```python
from scipy.stats import norm
def find_index(z,my_sol_cdf,starting_index):
    if my_sol_cdf[starting_index]>z:
        return starting_index
    for i in range(starting_index,len(my_sol_cdf)-1):
        if my_sol_cdf[i+1]>z and my_sol_cdf[i]<=z:
            break
    return i
def compute_W2_dist(new_meas_cdf,my_sol_cdf,new_xx):
    z_diff = np.array([new_meas_cdf[1]-new_meas_cdf[0]]+[new_meas_cdf[i+1]-new_meas_cdf[i] for i in range(len(new_meas_cdf)-1)])
    starting_index = 0
    dist = 0
    for j,z in enumerate(new_meas_cdf):
        starting_index = find_index(z=z,my_sol_cdf=my_sol_cdf,starting_index=starting_index)
        if starting_index == 0:
            cdf_inverse = new_xx[starting_index]
        else:
            ratio = (z-my_sol_cdf[starting_index])/(my_sol_cdf[starting_index+1]-my_sol_cdf[starting_index])
            if ratio<1:
                cdf_inverse = new_xx[starting_index] + ratio*(new_xx[starting_index+1]-new_xx[starting_index])
            else:
                cdf_inverse = new_xx[starting_index] + 1/2*(new_xx[starting_index+1]-new_xx[starting_index])
        #print(cdf_inverse,(z-my_sol_cdf[starting_index])/(my_sol_cdf[starting_index+1]-my_sol_cdf[starting_index]))
        dist += (new_xx[j] - cdf_inverse)**2 * z_diff[j]
    return dist**(1/2)

def transport_map(xx,my_sol_cdf,new_xx):
    if ref_dist == 1:
        ref_cdf = norm.cdf(xx)
    if ref_dist == 2:
        ref_cdf = (xx+1)/2
    if ref_dist == 3:
        ref_cdf = 1/2+1/2*(1-np.exp(-lam*np.abs(xx)))*np.sign(xx)
    if ref_dist == 4:
        upper_bound = np.minimum(xx,x_upper)
        ref_cdf = (xx>x_lower)*t0**(-1/3)*(C*(upper_bound-x_lower)-(upper_bound**3-x_lower**3)*t0**(-2/3)/36)
    
    t_map = np.zeros(len(xx))
    starting_index = 0
    for (i,z) in enumerate(ref_cdf):
        starting_index = find_index(z=z,my_sol_cdf=my_sol_cdf,starting_index=starting_index)
        if starting_index == 0:
            t_map[i] = new_xx[starting_index]
        else:
            ratio = (z-my_sol_cdf[starting_index])/(my_sol_cdf[starting_index+1]-my_sol_cdf[starting_index])
            if ratio<1:
                t_map[i] = new_xx[starting_index] + ratio*(new_xx[starting_index+1]-new_xx[starting_index])
            else:
                t_map[i] = new_xx[starting_index] + 1/2*(new_xx[starting_index+1]-new_xx[starting_index])
    return t_map

```
